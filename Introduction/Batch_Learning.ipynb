



### 📌 Introduction
The video discusses an **alternative way to classify Machine Learning (ML)** algorithms based on **how they are trained**, rather than how they are supervised (e.g., supervised, unsupervised, reinforcement learning).

This classification includes:
- **Batch Learning (Offline Learning)**
- **Online Learning**

---

## 📊 Production vs Development in ML

### 🔧 Development Environment:
- Where models are **built and trained** (usually on a local machine).
- Data scientists and ML engineers develop and test the model here.

### 🚀 Production Environment:
- The final, trained model is **deployed on a server**.
- The server is accessible via IP address for real users.
- This environment serves **real-time predictions or recommendations**.

---

## 1️⃣ **Batch Learning (a.k.a. Offline Learning)**

### 🧠 Definition:
Batch Learning is the **traditional method** of ML model training where:
- The entire dataset is used **at once** to train the model.
- No incremental or continuous learning occurs after deployment.

### 🔍 Key Characteristics:
- Model is trained **offline** using all the available historical data.
- Once trained, the model is **fixed (static)**.
- To improve or update, the model must be **re-trained from scratch** with new + old data.

### 📦 Process Overview:
1. **Gather data**
2. **Train model** with the full dataset
3. **Test the model**
4. **Deploy the model to production**
5. **Model remains fixed** until next manual update

### ⚠️ Limitation:
- The model **does not adapt** to new data or changes in user behavior until it's retrained.
- As time passes, the model might become **outdated** due to:
  - Changes in market conditions
  - New user trends
  - Evolving data distributions (concept drift)

### 📉 Disadvantages:
- Cannot adapt to real-time data.
- Re-training is **resource-intensive and time-consuming**.
- Difficult to handle **very large datasets**.
- High **cost of storage and computation** for retraining.
