

### üîπ 1. **What is AI (Artificial Intelligence)?**

**Definition:**
- AI refers to the simulation of human intelligence in machines.
- The goal is to create systems that can think, reason, and solve problems like humans.

**Key Concepts:**
- AI is the **broadest field**.
- It includes everything from rule-based systems to advanced learning algorithms.
- AI can be **symbolic (rule-based)** or **learning-based**.

**Example:**
- Voice assistants like Siri, Alexa.
- Recommendation systems on YouTube and Netflix.

**Types of AI:**
- **Narrow AI**: Performs a specific task (e.g., image recognition).
- **General AI**: Hypothetical ‚Äì would match human intelligence.
- **Super AI**: Surpasses human intelligence ‚Äì still a concept, not reality.

---

### üîπ 2. **What is ML (Machine Learning)?**

**Definition:**
- ML is a **subset of AI** that allows machines to learn from data without being explicitly programmed.

**Main Idea:**
- Instead of writing rules manually, machines learn **patterns** and **rules** from data.

**Key Characteristics:**
- You feed data and labels to the model.
- The model learns the underlying patterns.
- You don‚Äôt write explicit logic ‚Äì the model derives it.

**Real-Life Example:**
- Showing many dog and non-dog images to a system.
- It learns to distinguish between dogs and not-dogs on its own.

**Types of ML:**
- **Supervised Learning** (with labeled data)
- **Unsupervised Learning** (with unlabeled data)
- **Reinforcement Learning** (learning by trial and error)

---

### üîπ 3. **What is DL (Deep Learning)?**

**Definition:**
- DL is a **subset of ML** that uses **neural networks with multiple layers** (deep neural networks).

**Inspired By:**
- Structure of the human brain (neurons and synapses).

**Key Features:**
- Automatically extracts complex features from raw data.
- Best suited for large datasets and high-performance hardware (like GPUs).

**Applications:**
- Image recognition (e.g., face recognition)
- Voice recognition (e.g., Google Voice)
- Natural Language Processing (e.g., ChatGPT)

**Example:**
- Feeding thousands of labeled images of dogs.
- The system learns to classify based on features like shape, size, color, etc.

### üîÑ Evolution of AI:

1. **Early AI** ‚Äì Rule-based systems.
   - Known as **Expert Systems** (e.g., diagnosing a disease based on symptoms).

2. **Limitations of Expert Systems**:
   - Not scalable for real-world complexity.
   - Couldn‚Äôt handle uncertain, dynamic, or large datasets.

3. **Rise of Machine Learning**:
   - Data-driven systems.
   - Better adaptability and scalability.

4. **Emergence of Deep Learning**:
   - Solves more complex problems like image and speech recognition.
