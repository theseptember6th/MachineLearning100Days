Let's be **very clear** about this:

---

ðŸ”µ **Terminal Prompt vs Python Current Working Directory**

- **Terminal prompt** (what your terminal shows, like `MachineLearning100Days %`) is **where your shell is**.
- **Python `os.getcwd()`** is **where your Python process is**.

Normally, **they are the same** â€” *but not always*.

---

### â— If you open a file from VS Code or an IDE:

Sometimes **the terminal prompt** may show `MachineLearning100Days %`,  
**but** when you run Python **from inside an editor**,  
the *editor* might set a different "working directory" behind the scenes.

**Example:**
- You open `working_with_csv_files/your_script.py` and click "Run".
- Even though the terminal says `MachineLearning100Days`, the *editor* runs Python **as if** it is inside `working_with_csv_files/`.
- So `os.getcwd()` gives you `/Users/kristalshrestha/Code/MachineLearning100Days/Gathering_Data/working_with_csv_files`.

---

### âœ… That's why:

- **If you run manually from Terminal** (with `python3 script.py`), then yes, the working directory is where you are in the terminal.
- **If you run from IDE button** ("Run" button in VS Code, PyCharm, etc.), it might **set a different working directory automatically** â€” depending on which file you are running.

**This is why `os.getcwd()` is the final truth**:  
> "Where Python *thinks* it is starting the program."

---

### ðŸ”¥ So to summarize for you:

| Situation | Where you run from | How to set path? |
|:---|:---|:---|
| Terminal says `MachineLearning100Days`, you run manually | relative to `MachineLearning100Days` |
| IDE/Editor runs it from deeper folder (like `working_with_csv_files`) | relative to `working_with_csv_files` |

âœ… Always use `os.getcwd()` to check before loading files.

---

### Solution for always working properly?

Use this instead of normal relative paths:
```python
import os
import pandas as pd

# Base directory is the directory of the current script file
base_dir = os.path.dirname(os.path.abspath(__file__))
csv_path = os.path.join(base_dir, 'aug_train.csv')

df = pd.read_csv(csv_path)
```

ðŸ”µ This way, **your code always knows where it is**, regardless of where you "Run" it from.

# **8.Squeeze Parameters**

````python
pd.read_csv('aug_train.csv',usecols=['gender'],squeeze=True)
````

Good â€” you're making great progress.  
Now your new error is:

```
TypeError: read_csv() got an unexpected keyword argument 'squeeze'
```

---

ðŸ”µ **Reason:**
- In **older versions of pandas** (`<1.4`), `squeeze=True` was allowed (to return a Series if possible).
- But in **newer pandas versions** (`>=1.4`), **`squeeze` was completely removed** from `read_csv()`.

That's why you see this error.

---

### âœ… **Correct way now (for new pandas):**

Instead of using `squeeze=True` inside `read_csv()`,  
you should first load it normally, then use `.squeeze()` separately if needed.

Example:

```python
import pandas as pd

# Step 1: Read only the 'gender' column
df = pd.read_csv('aug_train.csv', usecols=['gender'])

# Step 2: Squeeze to Series if you want
gender_series = df.squeeze()

print(gender_series)
```

---

### ðŸ”¥ Even better:
If you want a **one-liner**, you can do:

```python
gender_series = pd.read_csv('aug_train.csv', usecols=['gender']).squeeze()
```

âœ… This first reads it, and immediately squeezes it to a `Series`.

---

### ðŸ“¢ Why was `squeeze` removed from `read_csv()`?
- It made the behavior *less predictable*.
- Now pandas prefers a **clear** two-step approach: first read a `DataFrame`, then squeeze **explicitly** if needed.

---

# **10. Encoding Parameter**

````python
pd.read_csv('zomato.csv')
````

Alright â€” the error you're getting is:

```
UnicodeDecodeError: 'utf-8' codec can't decode byte 0xed in position 7044: invalid continuation byte
```

---

# ðŸ›  What it means:

- `pandas.read_csv()` **assumes** by **default** that your file is **UTF-8 encoded**.
- Your `zomato.csv` file **is NOT UTF-8** â€” it has some different encoding (like Latin-1 or Windows-1252 or something else).
- So when Pandas tries to decode it as UTF-8, **it crashes** at some bad byte (`0xed`).

---

# âœ… How to fix it:

Simply **specify the encoding manually** when reading:

```python
pd.read_csv('zomato.csv', encoding='latin1')
```
or sometimes
```python
pd.read_csv('zomato.csv', encoding='cp1252')
```
(depends slightly on how the CSV was saved).

---

# ðŸ“Œ Quick Rule:

| If error says | Try using encoding |
|:---|:---|
| `'utf-8' codec can't decode byte` | Try `'latin1'`, `'cp1252'`, `'ISO-8859-1'` |

---

# ðŸ“¢ So for you:

âœ… Correct way:

```python
df = pd.read_csv('zomato.csv', encoding='latin1')
```
(or if still error, try `'cp1252'`).

---

# ðŸ“¦ Why does this happen?

Because many real-world CSV files (especially downloaded from old websites, Excel, etc.)  
are saved in **Windows-1252** or **Latin-1** encoding instead of UTF-8.

## so as you can see we converted date column to date type

Once you convert a column (like `'date'`) to a **`datetime`** type using `parse_dates`, you unlock a variety of powerful operations and features that are not available when the column is a **string**. Hereâ€™s a breakdown of what you can **do now** with the `datetime` type and what you **couldn't do** easily with a string:

### **What You Can Do With a DateTime Column**:
1. **Extract Date Components**:
   - You can easily extract specific components of the date, such as:
     - **Year**: `df['date'].dt.year`
     - **Month**: `df['date'].dt.month`
     - **Day**: `df['date'].dt.day`
     - **Weekday**: `df['date'].dt.weekday` (0=Monday, 6=Sunday)
     - **Hour, Minute, Second**: `df['date'].dt.hour`, `df['date'].dt.minute`, `df['date'].dt.second`

2. **Date Arithmetic**:
   - You can subtract one date from another to get the **difference** in days, months, or years:
     - `df['date'] - pd.to_datetime('2020-01-01')` gives you the difference in days.
   
3. **Resampling and Time Series**:
   - If youâ€™re working with time series data, you can resample your data by different time periods like day, week, month, etc.
     - Example: `df.resample('M', on='date').sum()` â€“ sums the data per month.
   
4. **Sorting**:
   - Sorting data by dates is easier and more efficient with `datetime` objects than strings.
     - Example: `df.sort_values('date')` will sort the rows by date.

5. **Filtering**:
   - You can filter data based on specific date ranges:
     - Example: `df[df['date'] > '2020-01-01']` will give you all records after January 1, 2020.
   
6. **Handling Missing Dates**:
   - Handling `NaT` (Not a Time) is easier with `datetime` columns, as `NaT` represents missing or null date values.

7. **Time-based Operations**:
   - You can perform operations like rolling windows, calculating moving averages, etc., on the time-based data:
     - Example: `df['date'].rolling(window=5).mean()` (if the data is numeric) for a 5-day rolling mean.

8. **Date Formatting**:
   - You can easily **format** the date into strings for reports or presentations.
     - Example: `df['date'].dt.strftime('%Y-%m-%d')` converts the date into the format `YYYY-MM-DD`.

### **What Can't Be Done (or is Harder) with String Type**:
1. **Date Calculations**:  
   - String columns are not designed to perform operations like subtracting or adding dates. You can't easily find the number of days between two dates as you can with `datetime`.
   
2. **Date Comparison**:
   - Comparing dates is much easier with `datetime`. You can't perform logical operations (e.g., `df['date'] > '2020-01-01'`) on string-based dates without first converting them.

3. **Efficient Sorting**:
   - Sorting dates as strings can lead to incorrect results (e.g., lexicographical sorting), as `"2020-12-01"` will appear before `"2021-01-01"`. With `datetime`, sorting works in the intended chronological order.

4. **Time-based Grouping**:
   - Grouping data by specific time periods (like by month or by year) is much easier with `datetime` than with strings.

5. **Handling Time Zones**:
   - Time zone conversion and handling are much easier with `datetime`. You can set a timezone using `.dt.tz_localize()` and convert it to another timezone with `.dt.tz_convert()`.

### Example of Some Operations:
```python
# Extract the year
df['year'] = df['date'].dt.year

# Extract the month
df['month'] = df['date'].dt.month

# Calculate the difference in days from a specific date
df['days_since'] = df['date'] - pd.to_datetime('2020-01-01')

# Filter data for matches after January 1, 2020
df_filtered = df[df['date'] > '2020-01-01']

# Sort data by date
df_sorted = df.sort_values('date')

# Group by month and sum other columns
df_monthly = df.resample('M', on='date').sum()
```

---

### Summary:

When you convert a column to a **`datetime`** type, you gain access to **powerful date and time manipulations** that would be cumbersome or inefficient with string types. This includes easier filtering, sorting, extracting date components, performing arithmetic, and handling time series operations.
